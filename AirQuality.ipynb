```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('airquality.csv')
```

```python
# 1
print(data.head(10))
```

```python
# 2
print("Longitud del dataset:", len(data))
```

```python
# 3
print(data.info())
print(data.describe())
```

```python
# 4
max_temp = data['Temp'].max()
print("TempMax:", max_temp)
```

```python
# 5
max_temp_row = data[data['Temp'] == max_temp]
print("mes/dia temperatura maxima:", max_temp_row[['Month', 'Day']])
```

```python
# 6
plt.scatter(data['Wind'], data['Solar.R'])
plt.xlabel('Vent (Wind)')
plt.ylabel('radiacio solar Solar (Solar.R)')
plt.title('Vent i Solar.R')
plt.show()
```

```python
# 7
temp_per_month = data.groupby('Month')['Temp'].mean()
temp_per_month.plot(kind='bar')
plt.ylabel('Temperatura Mitjana')
plt.title('Temperatura Mitjana per Mes')
plt.show()
```

```python
# 8
plt.scatter(data['Ozone'], data['Temp'])
plt.xlabel('Ozo')
plt.ylabel('Temperatura')
plt.title('Oz i Temperatura')
plt.show()
```
