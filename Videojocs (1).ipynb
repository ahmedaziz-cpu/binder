```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('dato.csv')
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df.dtypes
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Genre', y='Global_Sales', data=df, estimator=sum)
plt.title('Vendes globals per gènere')
plt.ylabel('Vendes globals (milions)')
plt.show()
```

```python
regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
sales = [41.36, 28.96, 3.77, 8.45] 

plt.pie(sales, labels=regions, autopct='%1.1f%%', startangle=90, colors=['gold', 'skyblue', 'lightgreen', 'coral'])
plt.title('Distribución de Ventas por Región')
plt.show()
```

```python
category_counts = df['Genre'].value_counts()

category_counts.plot(kind='bar', color='skyblue')
plt.title('Conteo de Categorías')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(df['NA_Sales'], df['Global_Sales'], color='purple', alpha=0.7)
plt.title('Relación entre Ventas en Japón y Ventas Globales')
plt.xlabel('Ventas en Japón (Millones)')
plt.ylabel('Ventas Globales (Millones)')
plt.show()
```

```python
sns.boxplot(data=df, x='Platform', y='Global_Sales')
plt.title('Distribución de Ventas Globales por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Ventas Globales (Millones)')
plt.xticks(rotation=45)
plt.show()
```
